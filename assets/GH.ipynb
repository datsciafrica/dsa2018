### Relative Humidity and Temperature

According to a [random google search](https://www.quora.com/What-is-the-relationship-between-temperature-and-humidity-Directly-proportional-or-inversely-proportional-if-temperature-increases-decreases-does-humidity-decrease-increase),

> "For example, at 5C  air temperature, if the atmosphere is  "saturated"  and will hold no more molecules of water vapor, the relative humidity  is said to be 100%.   If the temperature of air raises by 10,   it will need about twice as many water molecules to reach saturation and the new relative humidity is 50%. If the temperature is  raised another 10, it will again double its capacity to hold water  vapor molecules. The relative humidity drops to 25% only." 

So do we see such a relationship in our measurements...
